```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
```

```python
driver = webdriver.Chrome("D:/chromedriver_win32\chromedriver.exe")
```

```python
driver.get('https://accommodationtimes.com/list-of-trade-association-all-over-india/')
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
name=[]
site=[]
tel=[]
address=[]
email=[]
p=soup.findAll('p')
# print(str(p[2]).split('>')[0])
for i in range(2,len(p)):
    name.append(str(p[i]).split('>')[1].split('<')[0])
    try:
        address.append(str(p[i]).split('>')[2].split('<')[0])
    except IndexError:
        address.append('.')
    try:
        tel.append((str(p[i]).split('>')[3].split('<')[0]))
    except IndexError:
        tel.append('.')
    try:
        email.append(str(p[i]).split('>')[4].split('<')[0].split('Email:')[1].split('Web')[0])
    except IndexError:    
        email.append('.')
    try:
        site.append(str(p[i]).split('>')[4].split('<')[0].split('Web:')[1])
    except IndexError:
        site.append('.')
# for i in range(2,len(p)):
#     print(p[i].)s
```

```python
a=pd.DataFrame('Name':name,'Address':address,'Phone':tel,'E-mail':email,'Website':site)
```

```python
df = pd.DataFrame(list(zip(name,address,tel,email,site)), 
columns =['Name', 'Address','Phone','E-Mail','Website']) 
```
