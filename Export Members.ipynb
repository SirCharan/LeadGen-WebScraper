```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
```

```python
driver = webdriver.Chrome("D:/chromedriver_win32\chromedriver.exe")
name=[]add1=[]add2=[]add3=[]email=[]
n='1'
for j in range(157):
    j=str(j)
    page='http://www.aepcindia.com/members/?pageno='+j
    driver.get(page)
    content = driver.page_source
    soup = BeautifulSoup(content)
    divs=driver.find_elements_by_xpath('//div[@class="{}"]'.format('views-row views-row-1 views-row-odd views-row-first'))
    for i in range(len(divs)):
        try:
            name.append(divs[i].text.split('\n')[0])
            email.append(divs[i].text.split('\n')[-1])
            add1.append(divs[i].text.split('\n')[1])
            add2.append(divs[i].text.split('\n')[2])
            add3.append(divs[i].text.split('\n')[3])
            print(i)
            print(email[i])
        except IndexError:
            break
        
```
