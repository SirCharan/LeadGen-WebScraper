```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
```

```python
drivers = webdriver.Chrome("D:/chromedriver_win32\chromedriver.exe")
```

```python
drivers.get('http://www.market-width.com/Associations/Banking-Financial-Systems-Associations-India.htm')
```

```python
content = drivers.page_source
soup = BeautifulSoup(content)
```

```python
name=[]
des=[]
website=[]
div=drivers.find_elements_by_xpath('//div[@class="{}"]'.format('ind_asscnlst_insidebox'))
for i in range(len(div)):
    a=str(div[i].text)
    name.append(a.split('\n')[0])
    des.append(a.split('\n')[-2])
    website.append(a.split('\n')[-1].split('Website:')[1])
    if name==des:
        des='.'
print(website)
```

```python
i=0
b=[]
c=[]
name=[]
des=[]
website=[]
a=soup.findAll('a')
for a in soup.findAll('a'):
#     print(i)
    b.append(str((a['href'])))
for i in range(10,70):
    c.append('http://www.market-width.com/Associations/'+b[i])
c
for i in range(len(c)):
    drivers.get(c[i])
    div=drivers.find_elements_by_xpath('//div[@class="{}"]'.format('ind_asscnlst_insidebox'))
    for i in range(len(div)):
        a=str(div[i].text)
        name.append(a.split('\n')[0])
        des.append(a.split('\n')[-2])
        website.append(a.split('\n')[-1].split('Website:')[1])
        if name==des:
            des='.'
print(website)
```

```python
df = pd.DataFrame(list(zip(name,des,website)), 
columns =['Name', 'Description','Website']) 
df.head()
```
